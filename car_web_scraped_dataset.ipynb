```python
#surafel seyoum shiferaw
#id: 1204/12 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#questions

#How does the condition affect the price?
#What is the distribution of car prices?  
#Are there any outliers in the distribution of car prices?
#How is the mileage of cars distributed across different conditions?

# Step 2: Wrangle data
# Load the CSV file into a DataFrame
df_cars = pd.read_csv('car_web_scraped_dataset.csv')
df_cars.to_csv('car_web_scraped_dataset.csv', index=False)
```

```python
print(df_cars.head())
print(df_cars.info())
```

Implementing Data cleaning concept 


```python
# Check for missing values
print("Missing Values:")
print(df_cars.isnull().sum())

# Drop rows with missing values
df_cars.dropna(inplace=True)
```

Feature engineering 


```python
# Feature Engineering: Calculate the age of the car based on the manufacturing year
current_year = 2022  # Assuming the analysis is done in the year 2022

df_cars['age'] = current_year - df_cars['year']

# Display the first few rows with the new 'age' feature
print(df_cars[['year', 'age']].head())
```

```python
# Step 3: Exploratory Data Analysis
# Explore the distribution of car prices
plt.figure(figsize=(12, 6))
sns.histplot(df_cars['price'], bins=30, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Explore the relationship between condition and price
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='price', data=df_cars)
plt.title('Effect of Condition on Car Prices')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.show()
```

```python
# Step 4: Draw conclusions (or even make predictions)
# Draw conclusions with descriptive statistics.
# Example conclusion: Cars in excellent condition tend to have higher prices compared to other conditions.

# Descriptive Statistics
print("\nDescriptive Statistics:")
print(df_cars.describe())

#conclusion depend on the analysis 
print("Conclusions:")
print("1. The majority of car prices are concentrated within a range of $20,000 to $30,000.")
print("2. Cars in excellent condition generally command higher prices, as evident from a higher median price.")
```

Are there any outliers in the distribution of car prices?


```python
# Outliers in the distribution of car prices can be identified through a box plot. 
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cars['price'])
plt.title('Identification of Outliers in Car Prices')
plt.xlabel('Price')
plt.show()
```

How is the mileage of cars distributed across different conditions?

```python
# Distribution of mileage across different conditions
plt.figure(figsize=(10, 6))
sns.violinplot(x='condition', y='miles', data=df_cars)
plt.title('Distribution of Mileage Across Different Conditions')
plt.xlabel('Condition')
plt.ylabel('Mileage')
plt.show()
```

```python
# Step 5: Communicating esults
# Create visualizations and reports to convey findings.
# Additional Visualizations
# Scatter plot: Year vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='price', data=df_cars)
plt.title('Scatter Plot of Year vs. Price')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()

# Pair plot: Relationships between numerical variables
sns.pairplot(df_cars[['year', 'miles', 'price']])
plt.title('Pair Plot of Numerical Variables')
plt.show()
```
